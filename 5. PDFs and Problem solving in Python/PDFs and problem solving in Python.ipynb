```python
import PyPDF2 as ppdf
import pdfminer as pdfm
import pdfminer.high_level
import pprint
```

```python
# mentioning the path
pdf = r'C:\Users\abin.varghese01\Documents\Data wrangling using Python\EN-FINAL Table 9.pdf'
```

```python
# making a file object
pdfFileObj = open(pdf, 'rb')
```

```python
# creating a pdf reader object
pdfReader = ppdf.PdfFileReader(pdfFileObj)
```

```python
# printing the number of pages in the pdf file
print(pdfReader.numPages)
```

```python
# creating a page object
pageObj = pdfReader.getPage(0)
```

```python
# extracting text from the page
print(pageObj.extractText())
```
